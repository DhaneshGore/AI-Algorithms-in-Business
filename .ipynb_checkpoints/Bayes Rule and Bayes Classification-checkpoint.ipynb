# Naive Bayes Classification - Simplified

Classification is a type of supervised learning where the goal is to assign a label to each data point. For example, living things could be classified as animals, plants, or fungi. Data points have features (like having feathers), which help predict their label.

In supervised learning, the algorithm first learns from a set of labeled data, then uses that knowledge to predict labels for new, unseen data.

## Naive Bayes Classification

Naive Bayes classification is a simple, fast method based on Bayes' Theorem. It assumes the features are independent, which makes it quick but still effective, especially when there are many f

### Bayes' Theorem
 are many features to analyze.


$$ P(L | \text{features}) = \frac{P(\text{features} | L)P(L)}{P(\text{features})} $$


The key idea is to use the probability of features given a label, $P(\text{features} | L)$, to predict the label given the features, $P(L | \text{features})$.

Bayes' Theorem tells us:

$$
P(L | \text{features}) = \frac{P(\text{features} | L)P(L)}{P(\text{features})}
$$

### Prediction Using Bayes' Theorem

We compare the posterior probabilities for different labels. If we have two labels, $L_1$ and $L_2$, we calculate the ratio of posterior probabilities to predict the label:

$\frac{P(L_1 | \text{features})}{P(L_2 | \text{features})} = \frac{P(\text{features} | L_1)P(L_1)}{P(\text{features} | L_2)P(L_2)}$

- If the ratio is greater than 1, assign label $L_1$.
- Otherwise, assign label $L_2$.

This simple approach is powerful, especially when there are many features to analyze.

We first demonstrate naive Bayes classification using Gaussian distributions.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import make_blobs
X,y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)
colors=np.array(["red", "blue"])
plt.scatter(X[:, 0], X[:, 1], c=colors[y], s=50)
for label, c in enumerate(colors):
    plt.scatter([], [], c=c, label=str(label))
plt.legend();
#plt.colorbar();
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
model = GaussianNB()
#model = MultinomialNB()
model.fit(X, y);
```

Naive Bayes algorithm fitted two 2-dimensional Gaussian distribution to the data. The means and the variances define these distributions completely.

```python
print("Means:", model.theta_)
print("Standard deviations:", model.var_)
```

## Accuracy
What is a situation when the accuracy score is not a good representation of the quality of a classification?

A. In cases where the dataset is unbalanced, the accuracy score might not be a useful indicator of classification quality. In unbalanced datasets, classes are distributed unevenly, with one class predominating the samples. In these situations, accuracy might be deceptive since a high accuracy score can be obtained by consistently predicting the majority class and ignoring the minority class. Consider evaluation metrics that take into account elements like precision, recall, F1 score, or area under the receiver operating characteristic curve (AUC-ROC), which offer insights into the performance of both classes, to obtain a more accurate evaluation of the classifier's performance.
